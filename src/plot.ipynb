```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data
data = {
    "strategy": [
        "closebook t5", "closebook Qwen", "t5", "t5+fewshot", "Qwen", "Qwen+fewshot",
        "closebook t5", "closebook Qwen", "t5", "t5+fewshot", "Qwen", "Qwen+fewshot",
        "closebook t5", "closebook Qwen", "t5", "t5+fewshot", "Qwen", "Qwen+fewshot",
    ],
    "metric": [
        "Answer Recall", "Answer Recall", "Answer Recall", "Answer Recall", "Answer Recall", "Answer Recall",
        "F1 Score", "F1 Score", "F1 Score", "F1 Score", "F1 Score", "F1 Score",
        "Exact Match", "Exact Match", "Exact Match", "Exact Match", "Exact Match", "Exact Match",
    ],
    "score": [
        14.71, 22.21, 29.36, 30.18, 39.75, 37.86,
        10.87, 22.76, 30.83, 31.34, 39.55, 39.44,
        0.71, 9.57, 15.78, 16.84, 19.50, 21.10,
    ]
}

df = pd.DataFrame(data)

# Use a low-saturation, low-contrast color palette
base_palette = sns.color_palette("light:#5A9", n_colors=6)

# Plot
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=df, x="metric", y="score", hue="strategy", palette=base_palette)

# Add value labels on top of each bar (skip zero bars)
for bar in ax.patches:
    height = bar.get_height()
    if height > 0:
        ax.annotate(
            f'{height:.2f}',
            xy=(bar.get_x() + bar.get_width() / 2, height),
            xytext=(0, 4),
            textcoords="offset points",
            ha='center',
            va='bottom',
            fontsize=9
        )

plt.title("Evaluation baseline and different generation models on test dataset", fontsize=14)
plt.ylabel("Score")
plt.xlabel("")
plt.ylim(0, 65)
plt.legend(title="strategy", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
